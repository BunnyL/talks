```python
import os
import pandas as pd
pd.set_option('notebook_repr_html', False)
pd.set_option('precision', 3)
```

```python
fpath = os.path.expanduser('~/Dropbox/data/fec.h5')
with pd.get_store(fpath) as store:
    frame = store['fec']
```

```python
frame
```

```python
frame.shape
```

Make the columns a little easier to read

```python
frame.columns = ['Candidate', 'Contributor', 
                 'City', 'State', 'Zipcode',
                 'Employer', 'Occupation',
                 'Amount', 'Date']
frame
```

Add party affiliations for each candidate using merge

```python
party = pd.read_csv('cand_af.csv', index_col=0, header=None, 
                    names=['Candidate', 'PoliticalParty'])
party
```

```python
joined = pd.merge(frame, party, left_on='Candidate', right_index=True)
```

```python
joined
```

### Time to Party

```python
rs = joined.groupby('PoliticalParty').sum().sort_index()
rs # notice how pandas excludes nuisance columns to start
```

```python
rs.plot(kind='bar') # Y-axis in units of $10mm
```

### Tête-à-Tête

```python
by_cand = joined.groupby('Candidate').sum().sort('Amount', ascending=False)
by_cand
```

### States

```python
states = pd.read_csv('state_id.csv', index_col=0, header=None, 
                    names=['State', 'StateID'])
states.tail(10)
```

```python
joined = pd.merge(joined, states, how='outer', left_on='State', right_index=True)
```

```python
by_state = joined.groupby('State').sum().sort('Amount', ascending=False)
by_state.head()
```

```python
by_state.head(25).plot(kind='bar')
```

```python
by_party_state = joined.groupby(['PoliticalParty', 'State']).sum()
by_party_state
```

```python
by_party_state.reset_index().sort(['PoliticalParty', 'State']).groupby('PoliticalParty').last()
```

Let's go through the pipeline step by step

```python
by_party_state
```

```python
stage1 = by_party_state.reset_index()
stage1
```

```python
stage1.head(10)
```

```python
stage2 = stage1.sort(['PoliticalParty', 'State'], ascending=False)
stage2
```
